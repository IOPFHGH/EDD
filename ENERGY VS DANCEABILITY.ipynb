```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\harsh\Downloads\spotify dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.track_name=df.track_name.fillna("VACUUM")
df.track_artist=df.track_artist.fillna("VACUUM")
df.track_album_name=df.track_album_name.fillna("VACUUM") 
```

```python
df.isnull().sum()
```

```python
df1=df.sort_values('track_popularity',ascending=False).head(1500)
df1
```

```python
plt.scatter(df1.energy,df1.danceability)
```

```python
from sklearn.cluster import KMeans
kmeans_model=KMeans(n_clusters=6)
kmeans_model.fit(df1[["energy","danceability"]])
```

```python
kmeans_model.cluster_centers_
```

```python
df1["cluster_group"]=kmeans_model.labels_
```

```python
#here we can see that which clusters are alloted to them
x=df1.iloc[:,11:]
x
```

```python
df1["cluster_group"].value_counts()
```

```python
import seaborn as sns
sns.scatterplot(x="energy",y="danceability",data=df1,hue="cluster_group")
```
